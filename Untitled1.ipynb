```python
import requests as rq
import json
import pandas as pd
```

```python
url="https://r.statista.com/wp-content/themes/statista-theme/addons/services/rankinglist.php?ranking_id=P-173224&topic_hub=healthcare"
```

```python
response=rq.get(url)
```

```python
response=response.json()
```

```python
response
```

```python
df=pd.DataFrame()
```

```python
rank=[]
name=[]
country=[]
link=[]
```

```python
response
```

```python
for i in range(0,50):
    rank.append(response['rankingList'][i]['rank'])
    name.append(response['rankingList'][i]['hospital']['name'])
    country.append(response['rankingList'][i]['hospital']['country'])
    link.append(response['rankingList'][i]['hospital']['link'])
```

```python
df['Rank']=rank
df['HospitalName']=name
df['Country']=country
df['Link']=link
```

```python
df.to_csv('top_50.csv')
```
